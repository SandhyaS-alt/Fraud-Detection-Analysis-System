# Exploratory Data Analysis(EDA) with Python:

### 1. Find the dimensions (number of rows and columns) of the dataset


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Loading dataset
data = pd.read_csv("/content/cc_data.csv")
# Dropping index column
data = data.drop(columns ='index')
# Dimensions of the Dataset
print("Dimensions of the Dataset:", data.shape)
```

### 2. Check if unique values are present in each categorical variable

```python
# Unique vqlues in each categorical variable
unique_values = data.select_dtypes(include = ['object']).nunique()
print("unique values in each categorical variable:")
print(unique_values)
```

### 3. Find the distribution of numerical variables in the dataset

```python
# Distribution of numerical variables
numerical_summary = data.describe()
print("Distribution of numerical variables:")
numerical_summary
```

### 4. Check if there is any missing values in the dataset

```python
# Missing values in the dataset
missing_values = data.isnull().sum()
print("Missing values in the dataset:")
print(missing_values)
```

```python
data = data.dropna()
```

### 5. Find the summary statistics (mean, median, min, and max) for numerical variables

```python
# Summary statistics for numerical variables
summary_statistics = data.describe(include='all')
print("Summary statistics for numerical variables:")
summary_statistics
```

### 6. Check if there is any correlation between numerical variables. If so, how strong is the correlation

```python
# Correlation between numerical variables
correlation_matrix = data.select_dtypes([float,int]).corr()
print("Correlation between numerical variables:")
correlation_matrix
```

### 7. Check if the distribution of an amt differ across is_fraud categories

```python
# How does the distribution of a amt differ across is_fraud categories?
sns.boxplot(x ='is_fraud', y='amt', data = data)
plt.show()
```

### 8. Check if there is any outliers in the city_pop and amt

```python
# Outliers in numerical variables
sns.boxplot(data = data[['amt']])
plt.show()

sns.boxplot(data = data[['city_pop']])
plt.show()
```

### 9. Check for trends or patterns in the data over time (if applicable)

```python
# Convert 'trans_date_trans_time' to datetime objects
data['trans_date_trans_time'] = pd.to_datetime(data['trans_date_trans_time'], format = '%d-%m-%Y %H:%M')

#Sort data by transaction date and time
data = data.sort_values(by = 'trans_date_trans_time')

#Plot transaction amount over time
plt.figure(figsize=(14,6))
plt.plot(data['trans_date_trans_time'],data['amt'])
plt.title('Transaction Amount Over Time')
plt.xlabel('Date and Time')
plt.ylabel('Amount')
plt.show()
```

### 10. How does the target variable (if available) distribute across different categories?

```python
# How does the target variable (if available) distribute across different categories?
category_fraud_distribution = data.groupby('category')['is_fraud'].value_counts().unstack(fill_value = 0)
print("Distribution of 'is_fraud' across different categories:")
display(category_fraud_distribution)
```

### 12. Are there any potential data entry errors or inconsistencies in the dataset?

```python
# Check for duplicate rows
duplicate_rows = data.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_rows}")

# Check data types of columns
print("\nData types of columns:")
print(data.dtypes)

# You can add more checks here based on domain knowledge, e.g.,
# checking for consistent formatting in date columns, or
# checking if certain values fall within a reasonable range.
```

### 13. How does the distribution of numerical variables vary between different groups or segments of the dataset?

```python
# Done by grouping the data by a categorical variable and then describing the numerical variables.
# For example, let's group by 'is_fraud' and look at the distribution of 'amt'

print("Distribution of 'amt' for fraudulent vs. non-fraudulent transactions:")
display(data.groupby('is_fraud')['amt'].describe())

# You can repeat this for other numerical variables and categorical groups of interest
```

### 14. What are the top factors that influence the target variable, if applicable?

```python
# We can look at correlations with the target variable
print("Correlation of numerical variables with 'is_fraud':")
display(data.select_dtypes([float, int]).corr()['is_fraud'].sort_values(ascending=False))

# For categorical variables, we can look at the distribution of 'is_fraud' within each category (already done in Task 10)
# We could also consider building a simple model to get feature importances, but for EDA, correlation is a good starting point.
```
